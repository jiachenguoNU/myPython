```python
filter_condition = (country_info['Population'] > 100) & (country_info['Area (sq. mi.)'] < 1000000)  # Define a filter condition
country_info.loc[filter_condition, 'Climate']  # Filter the DataFrame based on the condition
```

```python
country_info.loc[~filter_condition]  # Filter the DataFrame based on the condition
```

```python
country_info.reset_index(inplace=True)  # Reset the index to default integer index
```

```python
countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan']
in_filder = country_info['Country'].isin(countries)  # Check if the index is in the list of countries
country_info[in_filder].sort_index(ascending=True, inplace=True)  # Filter the DataFrame based on the condition
```
